### Exercice : Validation d'un mot de passe

1. Demander à l'utilisateur de saisir un mot de passe.  
2. Si le mot de passe contient moins de 6 caractères, afficher **"Mot de passe faible"**.  
3. Sinon, si le mot de passe contient plus de 6 caractères **et** inclut tous les caractères `@`, `#`, `!`, afficher **"Mot de passe fort"**.  
4. Si le mot de passe contient plus de 6 caractères mais ne contient pas tous ces caractères, afficher **"Mot de passe moyen"**.

```python
mot_de_passe= input("Entrez votre mot de passe : ")
x= '@'
y='#' 
z='!'
if len(mot_de_passe) < 7:
    print("Mot de passe trop court, il doit contenir au moins 6 caractères.")
elif x in mot_de_passe and y in mot_de_passe and z in mot_de_passe and len(mot_de_passe) > 7:
    print("Mot de passe fort")
else:
    print("Mot de passe valide.")
```

### Exercice : Somme et produit des nombres saisis

1. Permettre à l'utilisateur d'entrer des nombres un par un.  
2. L'utilisateur peut continuer à saisir des nombres autant de fois qu'il le souhaite.  
3. Lorsque l'utilisateur saisit **"exit"**, le programme s'arrête.  
4. Le programme affiche alors :  
   - La **somme** de tous les nombres saisis.  
   - La **multiplication** (produit) de tous les nombres saisis.


```python
la_somme = 0
multiplication=1
while True : 
    x=input('enter a number: ')
    if x== 'exit':
        print ('la somme; ' ,)
        print ('la multiplication' , )
        break
    else:
        x = int(x)
        multiplication *= x
        la_somme += x
        print ('la somme :' ,la_somme)
        print ('multiplication :' ,multiplication)
```

### Exercice : Résolution d'une équation quadratique

On considère l'équation :  

$$ax^2 + bx + c = 0$$

1. Demander à l'utilisateur de saisir les valeurs de `a`, `b`, et `c`.  
2. Calculer le discriminant :  

$$\Delta = b^2 - 4ac$$

3. Selon la valeur de $\Delta$ :  
   - Si $\Delta > 0$ : deux solutions réelles distinctes  

     $$x_1 = \frac{-b - \sqrt{\Delta}}{2a}, \quad x_2 = \frac{-b + \sqrt{\Delta}}{2a}$$

   - Si $\Delta = 0$ : une solution réelle double  

     $$x = \frac{-b}{2a}$$

   - Si $\Delta < 0$ : pas de solution réelle (solutions complexes).

```python
import math
print ("let A=ax^2+bx+c")
a=int(input('la valeur de a:'))
b=int(input("la valeur de b:"))
c=int(input('la valeur de c:'))
delta=b**2-4*a*c
print("la valeur de delta=" ,delta)
if delta>0 :
    print("deux solutions reelles distinctes")
    x1=(-b-math.sqrt(delta))/(2*a)
    x2= (-b+math.sqrt(delta))/(2*a)
    print ("x1",x1)
    print("x2:",x2)
elif delta==0:
    print("une solutions reelle double")
    x0 = -b/2*a
    print("x=" , x0)
else: 
    print ("pas de solution reelle (solutions complexes)")
```
