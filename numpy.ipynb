```python
import numpy as np
grades = np.array([85, 90, 88, 92, 95, 80, 75, 98, 89, 83])
mean_grade = np.mean(grades)
median_grade = np.median(grades)
std_deviation = np.std(grades)
max_grade = np.max(grades)
min_grade = np.min(grades)
sorted_grades = np.sort(grades)
highest_grade_index = np.argmax(grades)
students_above_90 = np.sum(grades > 90)
percentage_above_90 = np.mean(grades > 90) * 100
percentage_below_75 = np.mean(grades < 75) * 100
high_performers = grades[grades > 90]
passing_grades = grades[grades > 75]
print("--- NumPy Student Grades Analysis ---")
print(f"1. Original Grades: {grades}")
print(f"2. Mean: {mean_grade:.2f} | Median: {median_grade} | Std Dev: {std_deviation:.2f}")
print(f"3. Maximum Grade: {max_grade} | Minimum Grade: {min_grade}")
print(f"4. Sorted Grades: {sorted_grades}")
print(f"5. Index of Highest Grade: {highest_grade_index}")
print(f"6. Number of Students Above 90: {students_above_90}")
print(f"7. Percentage Above 90: {percentage_above_90:.1f}%")
print(f"8. Percentage Below 75: {percentage_below_75:.1f}%")
print(f"9. High Performers Array: {high_performers}")
print(f"10. Passing Grades Array: {passing_grades}")
```
